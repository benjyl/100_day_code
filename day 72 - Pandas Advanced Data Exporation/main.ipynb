```python
# import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
# Web scraping to get salaries of multipage table
more_pages = True
page = 1
data = []
column_names = []

while more_pages:
    # print(page)
    url = f"https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors/page/{page}"
    response = requests.get(url)

    response.encoding = "utf-8"
    top_salaries = response.text
    soup = BeautifulSoup(top_salaries, "html.parser")
    rows = soup.find_all(class_="data-table__row")
    
    # next = soup.find(class_ = "pagination__next-btn")
    # print(rows[0])
    for row in rows:
        row_data = []
        for cell in row.find_all("td"):
            cell = cell.text.split(":")
            cell_info = cell[1].replace('$', '').replace(',' , '')
            print(cell_info)
            if cell_info.isnumeric() and len(cell_info) > 3:
                cell_info = int(cell_info)
                cell[0] = cell[0] + " ($)" 
                print(cell_info)  
            row_data.append(cell_info)
            if cell[0] not in column_names:
                column_names.append(cell[0])
            
        # print(cell.text)
        data.append(row_data)
    
    if soup.select('a.pagination__next-btn.pagination__btn--off') == []:
        page +=1
    else:
        more_pages = False

```

```python
# store data in pandas dataframe

df = pd.DataFrame(data, columns=column_names)
# df.set_index('Rank', inplace=True)
df.head(57)
```
